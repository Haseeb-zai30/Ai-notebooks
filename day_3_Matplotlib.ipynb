**Matplotlib**

Matplotlib is a powerful plotting library in Python used for data visualization.

It allows us to create line plots, bar charts, histograms, scatter plots, and more.

pyplot is a module in matplotlib that provides functions similar to MATLAB for creating plots easily.

```python
import matplotlib.pyplot as plt
```

```python
# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
```

plt.plot(x, y) → Plots a line graph with x values on the horizontal axis and y values on the vertical axis.

plt.title() → Adds a title to the graph.

plt.xlabel() & plt.ylabel() → Labels for axes.

plt.show() → Displays the final plot (without this, sometimes plots won’t appear).

```python
# Add title and labels
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
# Create a basic line plot
plt.plot(x, y);# a plot between x and y
plt.show()
```

**2. Customizing Line Plots**

color="red" → Sets line color (can also use "g", "b", "black", or hex codes like "#ff5733").

linestyle="--" → Dashed line ("-" solid, ":" dotted, "-." dash-dot).

marker="o" → Marks each data point with circles ("s" square, "^" triangle, "*", etc).

linewidth=2 → Thickness of line.

markersize=8 → Size of markers.

plt.grid(True) → Adds a grid for better readability.

```python
# Customizing line: color, linestyle, marker
plt.plot(x, y, color="red", linestyle="--", marker="o", linewidth=2, markersize=8)
plt.grid(True)# Show grid
```

**3. Plotting Multiple Lines**

plt.legend() → Adds a box that explains which line belongs to which dataset.

If you don’t use label=, legend will be empty.

```python
# Data
x = [1, 2, 3, 4, 5]
y1 = [1, 2, 3, 4, 5]
y2 = [2, 4, 6, 8, 10]
y3 = [1, 4, 9, 16, 25]

# Plot multiple lines with labels
plt.plot(x, y1, label="y = x")
plt.plot(x, y2, label="y = 2x")
plt.plot(x, y3, label="y = x^2")
plt.legend()
plt.show()
```

**4. Changing Figure Size and Saving Plots:**

plt.figure(figsize=(8, 5)) → Creates a larger or smaller canvas (width=8 inches, height=5 inches).

plt.savefig("my_plot.png") → Saves the plot as an image (.png, .jpg, .pdf supported).

```python
# Adjust figure size
plt.figure(figsize=(8, 5))

x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

plt.plot(x, y, color="purple", marker="o")

plt.title("Figure Size & Save Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Save the plot as an image
plt.savefig("my_plot.png")  # saves in current directory

plt.show()
```

**Bar Plots:**

A bar plot is used to compare data between categories.

plt.bar(x, y) → Creates vertical bars where x are categories and y are values.

color → Fill color of bars.

edgecolor → Border color of bars.

```python
# Data
categories = ["A", "B", "C", "D", "E"]
values = [4, 7, 2, 9, 5]

# Create bar plot
plt.bar(categories, values, color="skyblue", edgecolor="black")

# Add labels and title
plt.title("Basic Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")

plt.show()
```

**Horizontal Bar Plot**

plt.barh() → Creates horizontal bars instead of vertical.

```python
# Horizontal bar plot
plt.barh(categories, values, color="lightgreen", edgecolor="black")

plt.title("Horizontal Bar Plot")
plt.xlabel("Values")
plt.ylabel("Categories")

plt.show()
```

**7. Comparing Multiple Bars (Grouped Bar Plot):**

np.arange(len(categories)) → Generates positions for bars.

width → Controls bar thickness.

x - width/2 and x + width/2 → Positions bars side by side.

plt.xticks() → Replaces numeric ticks with category names.

```python
import numpy as np

# Data
categories = ["A", "B", "C", "D"]
values1 = [3, 7, 5, 8]
values2 = [4, 6, 8, 6]

# Positions of bars
x = np.arange(len(categories))   # [0,1,2,3]
width = 0.35  # width of bars each bar takes up 35% of the unit space (1).This makes sure that bars from Set 1 and Set 2 don’t overlap too much.

plt.bar(x - width/2, values1, width, label="Set 1", color="orange")#x - width/2 → shifts the bar positions slightly to the left of each category index.
plt.bar(x + width/2, values2, width, label="Set 2", color="blue")#x +width/2 → shifts the bar positions slightly to the right of each category index.

# Add labels
plt.xticks(x, categories)  # Replace numbers with category names
plt.title("Grouped Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.legend()

plt.show()
```

**8. Histograms (Distribution of Data):**

Histograms show the frequency distribution of data.

Take your dataset data (which should be a 1D array/list of numerical values).

Break the range of values into 30 bins (intervals).

Count how many values fall into each bin.

Draw vertical bars representing those counts.

```python
# Generate random data
data = np.random.randn(1000)  # 1000 values from normal distribution

plt.hist(data, bins=30, color="skyblue", edgecolor="black")

plt.title("Histogram Example")
plt.xlabel("Value Ranges")
plt.ylabel("Frequency")
plt.show()
```

**9. Scatter Plot (Relationship Between Two Variables):**

Scatter plots show correlation between two variables.

plt.scatter(x, y) → Plots points instead of lines.

marker → Shape of points.

s → Size of points.

```python
# Random data
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color="red", marker="o", s=10)  # s = marker size

plt.title("Scatter Plot Example")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

**10. Pie Chart (Proportions):**
Creates a pie chart, which shows data as slices of a circle.

Each slice’s size is proportional to its value compared to the total.

  plt.pie(sizes, labels=labels) → Creates pie chart.

 **autopct="%1.1f%%"**

Displays the percentage value inside each slice.

"%1.1f%%" format means:

1.1f → one digit before decimal, one digit after decimal (e.g., 25.3).

%% → percent sign.

**startangle**=90

Rotates the pie chart so the first slice starts at 90° (top of the circle) instead of the default 0° (right side).

This is often used to make the chart look more balanced or standardized.


```python
# Data
labels = ["Apples", "Bananas", "Cherries", "Dates"]
sizes = [30, 25, 20, 25]

plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90, colors=["red", "yellow", "pink", "brown"])

plt.title("Pie Chart Example")
plt.show()
```

**Area Plot (Filled Line Plot):**

plt.fill_between() → Fills the area under the curve.

alpha=0.4 → Controls transparency.

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.fill_between(x, y, color="skyblue", alpha=0.4)  # alpha = transparency
plt.plot(x, y, color="blue")

plt.title("Area Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### Box Plot

Box plots (also called whisker plots) are used to show the distribution of data, highlighting:

Median (middle value)

Quartiles (25%, 75%)

Minimum & Maximum values

Outliers

```python
# Generate random data
np.random.seed(42)  # for reproducibility
data = np.random.normal(100, 15, 200)  # mean=100, std=15, 200 points

# Create boxplot
plt.boxplot(data)

plt.title("Basic Box Plot")
plt.ylabel("Values")
plt.show()
```

Multiple Box Plots

```python
# Create multiple datasets
data1 = np.random.normal(100, 10, 200)# mean=100, std=10, 200 points
data2 = np.random.normal(90, 20, 200)# mean=90, std=20, 200 points
data3 = np.random.normal(110, 5, 200)# mean=110, std=5, 200 points

# Boxplot with multiple datasets
plt.boxplot([data1, data2, data3], labels=["Group 1", "Group 2", "Group 3"])

plt.title("Multiple Box Plots")
plt.ylabel("Values")
plt.show()
```

### **Subplots (Multiple Plots in One Figure):**
### a) Simple 1 Row, 2 Columns
plt.subplot(r, c, n) → Creates a grid of plots with r rows and c columns, and selects plot number n.

plt.tight_layout() → Adjusts spacing so titles/labels don’t overlap.

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 100)  # 100 points between 0 and 10
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
# Create 1 row, 2 columns of subplots
plt.subplot(1, 2, 1)   # (rows, cols, plot number)
plt.plot(x, y1, color="blue")
plt.title("Sine Wave")
plt.subplot(1, 2, 2)
plt.plot(x, y2, color="red")
plt.title("Cosine Wave")

plt.tight_layout()  # Avoids overlapping
plt.show()
```

### b) 2 Rows, 2 Columns

```python
x = np.linspace(0, 10, 100)# 100 points between 0 and 10
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(-x)

plt.subplot(2, 2, 1)# (rows, cols, plot number)
plt.plot(x, y1, "b")
plt.title("Sine")

plt.subplot(2, 2, 2)# (rows, cols, plot number)
plt.plot(x, y2, "r")
plt.title("Cosine")

plt.subplot(2, 2, 3)# (rows, cols, plot number)
plt.plot(x, y3, "g")
plt.title("Tangent")

plt.subplot(2, 2, 4)# (rows, cols, plot number)
plt.plot(x, y4, "m")
plt.title("Exponential Decay")

plt.tight_layout()
plt.show()
```

### c) Using plt.subplots() (Better Method):


Gives you fig (whole figure) and ax (individual axes).

More control: each ax can be customized separately.

```python
# Create a 2x2 grid
fig, ax = plt.subplots(2, 2, figsize=(8, 6))

# Top-left
ax[0, 0].plot(x, y1, "b")
ax[0, 0].set_title("Sine")

# Top-right
ax[0, 1].plot(x, y2, "r")
ax[0, 1].set_title("Cosine")

# Bottom-left
ax[1, 0].plot(x, y3, "g")
ax[1, 0].set_title("Tangent")

# Bottom-right
ax[1, 1].plot(x, y4, "m")
ax[1, 1].set_title("Exponential")

plt.tight_layout()
plt.show()
```

## Practice on Dataset

### Step 1: Load the Dataset

Save your dataset as a CSV (e.g., vehicles.csv) and load it with pandas.

```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
url="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%202/data/FuelConsumptionCo2.csv"

response=requests.get(url)#download the dataset
file="vehicles.csv"
with open(file,"wb") as x: #open file in write mode
    x.write(response.content)# write content to the file

# Load dataset
df = pd.read_csv("vehicles.csv")

# Show first few rows
print(df.head())
```

### Step 2: Basic Visualization Examples
a) Scatter Plot — Engine Size vs CO2 Emissions

```python
plt.scatter(df["ENGINESIZE"], df["CO2EMISSIONS"], color="red", alpha=0.5)

plt.title("Engine Size vs CO2 Emissions")
plt.xlabel("Engine Size (L)")
plt.ylabel("CO2 Emissions (g/km)")

plt.show()
```

This shows whether larger engines produce more CO₂.

### b) Histogram — Fuel Consumption in City

```python
plt.hist(df["FUELCONSUMPTION_CITY"], bins=30, color="skyblue", edgecolor="black")

plt.title("Fuel Consumption in City")
plt.xlabel("L/100km")
plt.ylabel("Frequency")

plt.show()
```

Shows the distribution of city fuel consumption across vehicles.

### c) Bar Plot — Average CO₂ by Fuel Type

```python
# Group by FUELTYPE and calculate mean CO2 emissions
avg_co2 = df.groupby("FUELTYPE")["CO2EMISSIONS"].mean()

avg_co2.plot(kind="bar", color="green", edgecolor="black")

plt.title("Average CO₂ Emissions by Fuel Type")
plt.xlabel("Fuel Type")
plt.ylabel("Average CO₂ Emissions (g/km)")
plt.show()
```

Compares CO₂ emissions for Gasoline, Diesel, Hybrid, etc.

### d) Box Plot — Cylinders vs CO₂ Emissions

```python
df.boxplot(column="CO2EMISSIONS", by="CYLINDERS", grid=False)

plt.title("CO₂ Emissions by Number of Cylinders")
plt.suptitle("")  # Remove extra title
plt.xlabel("Cylinders")
plt.ylabel("CO₂ Emissions (g/km)")
plt.show()
```

Helps you see variation/outliers in emissions for different cylinder counts.

### e) Multiple Line Plot — City vs Highway Consumption

```python
x = df["ENGINESIZE"]
y1 = df["FUELCONSUMPTION_CITY"]
y2 = df["FUELCONSUMPTION_HWY"]

plt.plot(x, y1, "r.", label="City Consumption")
plt.plot(x, y2, "b.", label="Highway Consumption")

plt.title("Fuel Consumption: City vs Highway")
plt.xlabel("Engine Size (L)")
plt.ylabel("Fuel Consumption (L/100km)")
plt.legend()
plt.show()
```

Compares how engine size affects consumption in city vs highway.
